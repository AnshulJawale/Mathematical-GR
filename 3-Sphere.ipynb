# Killing Vectors of a 3-Sphere ($S^3$)

The metric of $S^3$ in polar coordinates $\theta, \phi, \varphi$ is given by 
$$g = d\varphi^2 + sin^2(\varphi) \left(d\theta^2 + sin^2(\theta)d\phi^2\right)$$

```python
%display latex
```

```python
version()
```

```python
M = Manifold(3, "S^3")
M
```

```python
X.<vph,th,ph> = M.chart(r"vph:(0,pi):\varphi th:(0,pi):\theta ph:(0,2*pi):\phi")
X
```

```python
g = M.metric("g")
g[0,0] = 1
g[1,1] = sin(vph)^2
g[2,2] = sin(vph)^2 * sin(th)^2
g[:]
```

```python
nabla = g.connection()
```

```python
xi = M.vector_field(r'\xi')
xi[0] = function('xi_1')(vph, th, ph)
xi[1] = function('xi_2')(vph, th, ph)
xi[2] = function('xi_3')(vph, th, ph)
xi.display()
```

```python
killing_eqs = nabla(xi.down(g)).symmetrize()
killing_eqs.display()
```

```python
killing_eqs.display_comp(only_nonredundant=True)
```

Solving these, we get
- $K_{1} = \cos\psi\,\cos\theta\,\frac{\partial}{\partial\psi} - \sin\psi\,\sin\theta\,\frac{\partial}{\partial\theta}$
- $K_{2} = \cos\psi\,\sin\theta\,\cos\phi\,\frac{\partial}{\partial\psi} + \sin\psi\,\cos\theta\,\cos\phi\,\frac{\partial}{\partial\theta} - \sin\psi\,\sin\phi\,\frac{\partial}{\partial\phi} $
- $K_{3} = \cos\psi\,\sin\theta\,\sin\phi\,\frac{\partial}{\partial\psi} + \sin\psi\,\cos\theta\,\sin\phi\,\frac{\partial}{\partial\theta} + \sin\psi\,\cos\phi\,\frac{\partial}{\partial\phi}$
- $K_{4} = -\sin\theta\,\cos\phi\,\frac{\partial}{\partial\theta}- \cos\theta\,\sin\phi\,\frac{\partial}{\partial\phi}$
- $K_{5} = -\sin\theta\,\sin\phi\,\frac{\partial}{\partial\theta} + \cos\theta\,\cos\phi\,\frac{\partial}{\partial\phi}$
- $K_{6} = -\sin\phi\,\frac{\partial}{\partial\phi}$
