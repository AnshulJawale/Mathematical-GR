```python
M = Manifold(4, r'\mathcal{M}', structure='Lorentzian')
M
```

```python
X.<t,r,th,ph> = M.chart(r"t r:(0,+oo) th:(0,pi):\theta ph:(0,2*pi):\phi")
X
```

```python
var('m')
assume(m>=0)
```

```python
g = M.metric()
g[0,0], g[1,1] = -(1-2*m/r), 1/(1-2*m/r)
g[2,2], g[3,3] = r^2, (r*sin(th))^2
g.display()
```

```python
g.christoffel_symbols_display(only_nonredundant=False)
```

```python
Riem = g.riemann()
Riem.display_comp(only_nonredundant=False)
Riem.display()
```

```python
C = g.weyl()
C.display()
```

```python
Ric = g.ricci()
Ric[:]
```

```python
R = g.ricci_scalar()
R.display()
```
