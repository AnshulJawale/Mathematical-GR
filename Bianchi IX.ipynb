```python
from sage.manifolds.all import *

M = Manifold(4, 'M', structure='Lorentzian')

X.<t, chi, theta, phi> = M.chart(
    r't:(-oo,oo) x y z'
)
```

```python
dt, dchi, dtheta, dphi = X.coframe()

omega1 = cos(phi)*dtheta + sin(phi)*sin(theta)*dchi
omega2 = -sin(phi)*dtheta + cos(phi)*sin(theta)*dchi
omega3 = dphi + cos(theta)*dchi
```

```python
a1 = function('a1')(t)
a2 = function('a2')(t)
a3 = function('a3')(t)
```

```python
g = M.metric('g')

# Time component
g[0,0] = -1
g[1,1] = a1^2
g[2,2] = a2^2
g[3,3] = a3^2

'''# omega1^2
g[2,2] += a1^2 * cos(phi)^2
g[1,1] += a1^2 * sin(phi)^2 * sin(theta)^2
g[1,2] += a1^2 * sin(phi)*cos(phi)*sin(theta)

# omega2^2
g[2,2] += a2^2 * sin(phi)^2
g[1,1] += a2^2 * cos(phi)^2 * sin(theta)^2
g[1,2] += -a2^2 * sin(phi)*cos(phi)*sin(theta)

# omega3^2
g[3,3] += a3^2
g[1,3] += a3^2 * cos(theta)
g[1,1] += a3^2 * cos(theta)^2'''
```

```python
Ric = g.ricci()
R = g.ricci_scalar()
Ein = Ric - 0.5 * R * g
```

```python
rho = function('rho')(t)
p = function('p')(t)

T = M.tensor_field(0,2,'T')
T[0,0] = rho
T[1,1] = p
T[2,2] = p
T[3,3] = p
T[:]
```

```python
EFE = M.tensor_field(0,2,'EFE')
EFE = Ein - T

EFE.display()
```
