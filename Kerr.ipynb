```python
M = Manifold(4, r'\mathcal{M}', structure='Lorentzian')
M
```

```python
X.<t,r,th,ph> = M.chart(r"t r:(0,+oo) th:(0,pi):\theta ph:(0,2*pi):\phi")
X
```

```python
var('m, a', domain='real')
rho2 = r^2 + (a*cos(th))^2
Delta = r^2 -2*m*r + a^2
```

```python
g = M.metric()
g[0,0] = -(1-2*m*r/rho2)
g[0,3] = -2*a*m*r*sin(th)^2/rho2
g[1,1], g[2,2] = rho2/Delta, rho2
g[3,3] = (r^2+a^2+2*m*r*(a*sin(th))^2/rho2)*sin(th)^2
g.display()
```

```python
g[:]
```

```python
g.christoffel_symbols_display(only_nonredundant=True)
```

```python
Riem = g.riemann()
Riem.display()
```

```python
C = g.weyl()
C.display()
```

```python
Ric = g.ricci()
Ric[:]
```

```python
R = g.ricci_scalar()
R.display()
```
