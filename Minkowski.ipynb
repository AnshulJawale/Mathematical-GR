```python
M = Manifold(4, r'\mathcal{M}', structure='Lorentzian')
M
```

```python
X.<t,r,th,ph> = M.chart(r"t x y z")
X
```

```python
g = M.metric()
g[0,0] = -1
g[1,1] = 1
g[2,2] = 1
g[3,3] = 1
g.display()
```

```python
g.christoffel_symbols_display(only_nonzero=False, only_nonredundant=True)
```

```python
Riem = g.riemann()
Riem.display_comp(only_nonredundant=False)
Riem.display()
```

```python
C = g.weyl()
C.display()
```

```python
Ric = g.ricci()
Ric[:]
```

```python
R = g.ricci_scalar()
R.display()
```
